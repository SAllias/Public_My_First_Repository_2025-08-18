```python
from datetime import datetime, timedelta
```

```python
df_new_date = df_data_from_file.copy()
```

```python
df_new_date.head(1)
```

```python
df_new_date['Date'] = pd.to_datetime(df_new_date['Date'], errors='coerce', utc=True)
```

```python
print(df_new_date['Date'].dtype)
```

```python
filter_df_25y_10m = df_new_date[(df_new_date['Date'].dt.year == 2025) & (df_new_date['Date'].dt.month == 10)]
filter_df_25y_10m.shape
```

```python
filter_df_25y_10m_plus_cl = filter_df_25y_10m.copy()
filter_df_25y_10m_plus_cl['Date_y_m_d'] = filter_df_25y_10m_plus_cl['Date']
```

```python
filter_df_25y_10m_plus_cl.head(1)
```

```python
filter_df_25y_10m_plus_cl['Date_y_m_d'] = filter_df_25y_10m_plus_cl['Date_y_m_d'].dt.strftime('%Y-%m-%d')
filter_df_25y_10m_plus_cl.head(1)
```

```python
info_df_25y_10m = pd.DataFrame({
    'Минимальная дата' : [filter_df_25y_10m_plus_cl['Date_y_m_d'].min()],
    'Максимальная дата' : [filter_df_25y_10m_plus_cl['Date_y_m_d'].max()],
    'Стоимость во время закрытия' : [filter_df_25y_10m_plus_cl.loc[
        filter_df_25y_10m_plus_cl['Date_y_m_d']
        == filter_df_25y_10m_plus_cl['Date_y_m_d'].max(),
        'Close'].iloc[0].round(2)]
}).T
print(info_df_25y_10m)
```
