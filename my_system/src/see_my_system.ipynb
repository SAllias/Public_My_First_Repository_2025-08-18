```python
import platform # будем смотреть с помощью него данные о macOS
```

```python
# Создаем словарь 
system_info = {
    "System" : platform.system(),
    "Release" : platform.release(),
    "Version" : platform.version(),
    "Machine" : platform.machine(),
    "Processor" : platform.processor()
}
# через цикл for выводим по строчно весь словарь
for key, value in system_info.items():
    print(f"{key} : {value}")
```

```python
import pandas as pd #положим словарь в таблицу
```

```python
# создаем базу, только вот она ключи преобразует в столбцы. По этому, добавил функцию транспорнирования
df = pd.DataFrame([system_info])
df_t = df.T.rename(columns={0:'Поле1'})
print(df)
print('\n', df.T)
print('\n', df_t)
print('\n', df_t.loc['Version'])
version_list_one = list(df_t.loc['Version'])
version_list_two = list(df_t.loc['Version',:])
version_list_three = list(df_t.loc['Version','Поле1'])
print('\nversion_list_one: ', version_list_one)
print('\nversion_list_two: ', version_list_two)
print('\nversion_list_three: ', version_list_three)
```
