```python
import yfinance as yf
import pandas as pd
import time

period_setup = '2y'  # –ü–µ—Ä–∏–æ–¥—ã 6mo(–º–µ—Å—è—Ü), 2y(–≥–æ–¥–∞)

def get_historical_data_fast(symbol, period=period_setup):
    try:
        print(f"üîÑ –ó–∞–≥—Ä—É–∑–∫–∞ {symbol} –∑–∞ {period}...")
        ticker = yf.Ticker(symbol)
        df = ticker.history(period=period)
        print(f"‚úÖ –ó–∞–≥—Ä—É–∂–µ–Ω–æ {len(df)} —Å—Ç—Ä–æ–∫")
        return df
    except Exception as e:
        print(f"‚ùå –û—à–∏–±–∫–∞: {e}")
        return None
```

```python
# –ò—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ
if __name__ == "__main__":
    print("üöÄ –ó–∞–ø—É—Å–∫ –±—ã—Å—Ç—Ä–æ–π –≤–µ—Ä—Å–∏–∏...")
    start_time = time.time()
    
    df = get_historical_data_fast('TSLA', period_setup) # –ù–∞–ø—Ä–∏–º–µ—Ä –∞–∫—Ü–∏–∏ AAPL, —Å–∞–π—Ç https://finance.yahoo.com. 
    
    if df is not None:
        print(df.head())
        print(f"\n–ü–æ—Å–ª–µ–¥–Ω—è—è —Ü–µ–Ω–∞: ${df['Close'].iloc[-1]:.2f}")
        print(f"‚è±Ô∏è –í—ã–ø–æ–ª–Ω–µ–Ω–æ –∑–∞: {time.time() - start_time:.2f} —Å–µ–∫—É–Ω–¥")
    else:
        print("–ù–µ —É–¥–∞–ª–æ—Å—å –∑–∞–≥—Ä—É–∑–∏—Ç—å –¥–∞–Ω–Ω—ã–µ")
```
